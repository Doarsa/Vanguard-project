# Distribution of Client Ages

```python
fig_age = px.histogram(df_exp_demo, x='clnt_age', nbins=30, title='Age Distribution of Clients')
fig_age.update_layout(xaxis_title='Age', yaxis_title='Count')
fig_age.show()
```

# Account Balance Distribution

```python
fig_raw = px.histogram(df_exp_demo, x='bal', nbins=50, 
                       title='Raw Account Balance Distribution')
fig_raw.update_layout(xaxis_title='Balance', yaxis_title='Client Count')
fig_raw.show()
```

The variable bal (account balance) is extremely skewed, meaning a small number of clients have very high balances while most have much lower ones. This distorts histograms and statistical models.

To fix this, we applied a logarithmic transformation using log1p (log(1 + x)), which compresses large values and reveals the distribution's true shape more clearly, making visualizations and insights more meaningful.

```python
df_exp_demo.loc[:, 'log_bal'] = np.log1p(df_exp_demo['bal'])
fig_bal = px.histogram(df_exp_demo, x='log_bal', nbins=30, 
                       title='Log-Scaled Distribution of Account Balance')
fig_bal.update_layout(xaxis_title='Log(Balance + 1)', yaxis_title='Count')
fig_bal.show()
```

# Number of Calls vs. Logins
